```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

import openhouse.etl.data_loader as dl
import openhouse.eda.eda_engine as eda

from pathlib import Path

EDA_FIG_SIZE = (3,3)
images_path = Path("img")
```

# 1.Data Loading <a class="data-loading"></a>

```python
# Import the dataset using the DataLoader class, which wraps several file types into one single data reader.
data_loader = dl.DataLoader("data/clean_dataset.csv")
clean_df = data_loader.load_data()
```

### <ins> 2. Explore the clean dataset<ins>
Based on: 
- the analysis of the data from the end of "02_Clean_Raw_Data"
- the amount of unique values
- Common sense regarding the nature of each feature

We are able to determine if a column is numerical ordinal or categorical.

### <ins> 2.1 Categorize each feature as numerical, ordinal or categorical<ins>
- There are 21 columns, but the last one is the target (SalePrice)<br>

- The numerical columns are five:
    1. LotAreaGrLivArea
    2. GrLivArea
    3. YearBuilt
    4. TotalBsmtSF
    5. GarageArea
<br>

- The ordinal columns are six:
    6. OverallQuality
    7. OverallCondition
    8. FullBath
    9. HalfBath
    10. GarageCars
    11. YearSold
<br>

- The categorical columns are nine:
    12. Street
    13. LotType
    14. BldgType
    15. HouseStyle
    16. Foundation
    17. CentralAir
    18. GarageType
    19. SaleType
    20. SaleCondition


```python
contin_num_cols = ['LotArea', 'GrLivArea', 'YearBuilt', 'TotalBsmtSF', 'GarageArea']
ordinal_cols = ['OverallQuality','OverallCondition','FullBath','HalfBath','GarageCars','YearSold']
categorical_columns = ['Street', 'LotType', 'BldgType', 'HouseStyle', 'Foundation', 'CentralAir', 'GarageType', 'SaleType', 'SaleCondition']
print(f"Numerical Columns: {contin_num_cols}")
print(f"Ordinal Columns: {ordinal_cols}")
print(f"Categorical Columns: {categorical_columns}")
```

```python
df_numerical = clean_df[contin_num_cols].copy()
df_ordinal = clean_df[ordinal_cols].copy()
df_categorical = clean_df[categorical_columns].copy()
y_df = clean_df['SalePrice'].copy()
```

### 2.2 <ins> Inspect each feature and its relationship with the target

### 2.2.1 Continous Numerical Features

```python
df = df_numerical.copy()
df['price'] = y_df.to_numpy()
num_imgs_path = images_path / "Numerical_Features"
os.makedirs(num_imgs_path, exist_ok=True)
for feat_name in contin_num_cols:
    fig = plt.figure(figsize = EDA_FIG_SIZE)
    ax = fig.subplots()
    sns.scatterplot(data=df, x=feat_name,y='price', hue='price', ax=ax);
    plt.xlabel("")
    plt.xticks(rotation=90);
    plt.title(feat_name, x=0.5, y=0.9)
    ax.get_legend().remove()
    img_fpath = num_imgs_path / feat_name
    plt.savefig(img_fpath)
```

```python
# plotting correlation heatmap
plt.figure(figsize=EDA_FIG_SIZE)
sns.heatmap(data=df.drop(columns=['price']).corr(), cmap="YlGnBu", annot=False)
img_fpath = num_imgs_path / "Numerical_Features_Correlation"
plt.savefig(img_fpath)
```

### 2.2.2 Ordinal Features

```python
df = df_ordinal.copy()
df['price'] = y_df.to_numpy()
ordinal_imgs_path = images_path / "Ordinal_Features"
os.makedirs(ordinal_imgs_path, exist_ok=True)
for feat_name in ordinal_cols:
    fig = plt.figure(figsize = EDA_FIG_SIZE)
    ax = fig.subplots()
    sns.boxplot(data=df, x=feat_name,y='price', hue=feat_name, ax=ax);
    plt.xlabel("")
    plt.xticks(rotation=90);
    plt.title(feat_name, x=0.5, y=0.9)
    img_fpath = ordinal_imgs_path / feat_name
    plt.savefig(img_fpath)
```

```python
# plotting correlation heatmap
plt.figure(figsize=EDA_FIG_SIZE)
sns.heatmap(data=df.drop(columns=['price']).corr(), cmap="YlGnBu", annot=False)
img_fpath = ordinal_imgs_path / "Ordinal_Features_Correlation"
plt.savefig(img_fpath)
```

### 2.2.3 Categorical Features

```python
df = df_categorical.copy()
df['price'] = y_df.to_numpy()
categ_imgs_path = images_path / "Categorical_Features"
os.makedirs(categ_imgs_path, exist_ok=True)
for feat_name in categorical_columns:
    fig = plt.figure(figsize = EDA_FIG_SIZE)
    ax = fig.subplots()
    sns.boxplot(data=df, x=feat_name,y='price', hue=feat_name, ax=ax);
    plt.xlabel("")
    plt.xticks(rotation=90);
    plt.title(feat_name, x=0.5, y=0.9)
    img_fpath = categ_imgs_path / feat_name
    plt.savefig(img_fpath)
```
